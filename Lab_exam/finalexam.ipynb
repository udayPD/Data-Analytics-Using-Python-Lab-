```python
titles_count_by_service = data[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()
print("Number of titles available on each streaming service:\n", titles_count_by_service)
```

```python
highest_titles_service = titles_count_by_service.idxmax()
print("Streaming service with the highest number of titles:", highest_titles_service)
```

```python
average_imdb_by_service = data.groupby(['Netflix', 'Hulu', 'Prime Video', 'Disney+'])['IMDb'].mean()
print("Average IMDb rating for titles available on each streaming service:\n", average_imdb_by_service)
```

```python
#HIGHEST IMDB SERVICE
highest_imdb_service = average_imdb_by_service.idxmax()
print("Streaming service with the highest average IMDb rating:", highest_imdb_service)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='IMDb', y='Netflix', data=data)
plt.title("Distribution of IMDb ratings on Netflix")
plt.show()
```
